```python
%%sql
create table if NOT EXISTS accounts_scd
(
  account_id int,
  customer_id int,
  account_type string,
  balance float,
  hash_key BIGINT,
  created_by STRING,
  created_date TIMESTAMP,
  updated_by STRING,
  updated_date TIMESTAMP
)
USING DELTA
LOCATION 'Tables/Gold/Account_SCD'

```

```python
%%sql
create table if NOT EXISTS customers_scd
(
  customer_id int,
  first_name string,
  last_name string,
  address string,
  city string,
  state string,
  hash_key BIGINT,
  created_by STRING,
  created_date TIMESTAMP,
  updated_by STRING,
  updated_date TIMESTAMP
)
USING DELTA
LOCATION 'Tables/Gold/customers_SCD'
```

```python
%%sql
create table if NOT EXISTS loanpayments_scd
(
  payment_id int,
  loan_id int,
  payment_date DATE,
  payment_amount float,
  hash_key BIGINT,
  created_by STRING,
  created_date TIMESTAMP,
  updated_by STRING,
  updated_date TIMESTAMP
)
USING DELTA
LOCATION 'Tables/Gold/loan_payments_SCD'
```

```python
%%sql
create table if NOT EXISTS loans_scd
(
  loan_id int,
  customer_id int,
  loan_amount float,
  interest_rate float,
  loan_term INT,
  hash_key BIGINT,
  created_by STRING,
  created_date TIMESTAMP,
  updated_by STRING,
  updated_date TIMESTAMP
)
USING DELTA
LOCATION 'Tables/Gold/loans_SCD'
```

```python
%%sql
create table if NOT EXISTS transactions_scd
(
  transaction_id int,
  account_id int,
  transaction_date date,
  transaction_amount float,
  transaction_type STRING,
  hash_key BIGINT,
  created_by STRING,
  created_date TIMESTAMP,
  updated_by STRING,
  updated_date TIMESTAMP
)
USING DELTA
LOCATION 'Tables/Gold/transactions_SCD'
```

```python
from pyspark.sql.functions import col
dbtable1.alias("trg").merge(df_source1.alias("src"), "trg.account_id = src.account_id") \
    .whenMatchedUpdate(
        set={
            "trg.account_id":"src.account_id",
            "trg.customer_id":"src.customer_id",
            "trg.account_type":"src.account_type",
            "trg.balance":"src.balance",
            "trg.hash_key":"src.hashkey",
            "trg.updated_date":current_timestamp(),
            "trg.updated_by":lit("databricks_Updated")
})\
.whenNotMatchedInsert(
    values={
        "trg.account_id":"src.account_id",
        "trg.customer_id":"src.customer_id",
        "trg.account_type":"src.account_type",
        "trg.balance":"src.balance",
        "trg.hash_key":"src.hashkey",
        "trg.created_date":current_timestamp(),
        "trg.created_by":lit("databricks"),
        "trg.updated_date":current_timestamp(),
        "trg.updated_by":lit("databricks")}
        )\
.execute()
```
